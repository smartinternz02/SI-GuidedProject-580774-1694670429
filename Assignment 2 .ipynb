```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("House Price India.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
plt.scatter(df.index,df['Price'])
```

```python
sns.jointplot(x='number of bedrooms',y='Price',data = df)
```

```python
variables = ['Price', 'living area', 'number of bedrooms']
correlation_matrix = df[variables].corr()

plt.imshow(correlation_matrix, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(variables)), variables)
plt.yticks(range(len(variables)), variables)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df[['Built Year', 'number of bedrooms','Price']])
```
